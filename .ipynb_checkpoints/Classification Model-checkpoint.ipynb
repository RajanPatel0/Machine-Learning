```python
#Logistic Regression Algorithm: statistical method used for binary classification problem , where outcome is categorial variable

#Credit Card Fraud Detection:
#Predict whether a transaction is fraudulent based on various features of transactions. This is a binary classification problem where the target variables are fraud

#Data Dictionary:
#Transaction id
#Transaction Amount
#Transaction Time
#Merchant Category
#Customer Age
#Customer Gender
#Customer Income
#Transaction Location
#Previous Fraud Count
#Fraud: 0(no) or 1(yes)
```

```python
#Building a Model using Logistic Regression Algorithm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder  
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix #accuracy score is acuracy parameter used to evatluate, confusion matric used to plot the decisions
```

```python
#Load dataset from a csv file
df=pd.read_csv('CreditCardData.csv')
df   #display dataset
```

```python
#summary statistics
df.describe()
```

```python
df.info()
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Distribution of target variable
sns.countplot(x='Fraud', data=df)   #taking fraud column in x-axis & find it in dataframe(df)
plt.title('Distribution Of Fraudlent Transactions')
plt.show()
```

```python
#Box Plot for Transaction Amount by Fraud status
sns.boxplot(x='Fraud', y='Transaction Amount', data=df)
plt.title('Box Plot Of Transaction amount By Fraud Status')
plt.show()
```

```python
#Distribution Of Fraud Detection by fraud status
sns.histplot(data=df, x='Transaction Amount', hue='Fraud', multiple='stack', bins=30)
plt.title('Transaction Amount Distribution by Fraud Status')
plt.show()
```

```python
#correlation Matrix
nums_cols= list()

for column in df.columns:
    if df[column].dtype != object:
        nums_cols.append(column)   #means if object tye or text then not include it in nums_cols as it only takes numerical data &(new version show error)

correlation_matrix= df[nums_cols].corr()
sns.heatmap(correlation_matrix, annot= True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Data Processing: Encode Categorial variable [Processing data before fitting it in model]

label_encoders= {} #In label encoder- we use label for each category
for column in ['Merchant Category','Customer Gender', 'Transaction Location']:
    le=LabelEncoder()  #it's a class help i creating instance (le)
    df[column]= le.fit_transform(df[column]) #& use this instance to look at column & its values for transforming them via fit transform- it map & return to df[column]
    label_encoders[column]=le   #then adding that to dict initialized at top

#Features & Target
X=df.drop(['Transaction ID', 'Fraud'], axis=1) #axis=1 means performing via coulmns & rows for axis=0  && need to drop (these two's) cuzz required in y below 
y=df['Fraud']
```

```python
df.head()   #Now object text values got changed or can say encoded via label encoder & got transformed
```

```python
#Split dataset into training & testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
#Feature Scaling- Solving Problems of Difference in numeric values of few data columns in or out
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
#Logistic Regression- as dataset is ready

from sklearn.linear_model import LogisticRegression

logreg= LogisticRegression() #creating instance of LogisticRegression class
logreg.fit(X_train, y_train)  #them a method fit to fit train of both X,y
logreg_pred= logreg.predict(X_test) #now predicting on x_test dataset
logreg_accuracy= accuracy_score(y_test, logreg_pred)  #actual values are in y_test & predicted isin LOGREG_PRED
```

```python
#display accuracy & performance metrix
print(f'Logistic Regression Accuracy: {logreg_accuracy*100:.2f}%') #multiply with 100 for % and set 2decimal format(.2f) in String Literal
print('\nLogistic Regression Classification Report: ')
print(classification_report(y_test, logreg_pred)) #passing our actual y_test & predicting on it
```

```python
# --- Final Live Prediction (User Input) ---

print("\n🔍 Let's test a new transaction...")

# Example input (replace these with input() for real-time)
amount = float(input("Enter Transaction Amount (INR): "))
time = int(input("Enter Transaction Time (0–86400 sec): "))
category = input("Enter Merchant Category (e.g. Grocery, Travel, etc.): ")
age = int(input("Enter Customer Age: "))
gender = input("Enter Customer Gender (Male/Female/Other): ")
income = float(input("Enter Customer Income: "))
location = input("Enter Transaction Location (City): ")
prev_frauds = int(input("Enter Previous Fraud Count: "))

# Encode user input using saved encoders
category_encoded = label_encoders['Merchant Category'].transform([category])[0]
gender_encoded = label_encoders['Customer Gender'].transform([gender])[0]
location_encoded = label_encoders['Transaction Location'].transform([location])[0]
input_dict = {
    'Transaction Amount': [amount],
    'Transaction Time': [time],
    'Merchant Category': [category_encoded],
    'Customer Age': [age],
    'Customer Gender': [gender_encoded],
    'Customer Income': [income],
    'Transaction Location': [location_encoded],
    'Previous Fraud Count': [prev_frauds]
}
user_df = pd.DataFrame(input_dict)

# Scale the user input
user_scaled = scaler.transform(user_df)

# Predict
user_prediction = logreg.predict(user_scaled)[0]

# Output Result
print("\n🎯 Final Prediction:")
print("🚨 Fraudulent Transaction ❌" if user_prediction == 1 else "✅ Safe Transaction")
```

```python

#Building A Model Using KNN Algorithm
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#K-Nearest Neighbors
knn= KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred= knn.predict(X_test) #we use pred to get rediction on our x test set(known for this algo)
knn_accuracy= accuracy_score(y_test, knn_pred)
```

```python
#help(KNeighborsClassifier)   -- it gives complete details
```

```python
#Display accuracy and performance metrics
print(f'K-Nearest Neighbors Accuracy: {knn_accuracy*100:.2f}')
print(f'K-Nearest Neighbors Classification Report:')
print(classification_report(y_test, knn_pred))
```

```python
#Confusion Metrix
plt.figure(figsize=(7,5))
sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True, fmt='d', cmap='Blues')
plt.title('K-Nearest Neighbors Confusion Matrix')
plt.xlabel('Actual +ve')
plt.ylabel('Predicted +ve')
plt.show()
```

```python
print("\n🔍 Test a New Transaction Using KNN Model...")

#  User Input
amount = float(input("Enter Transaction Amount (INR): "))
time = int(input("Enter Transaction Time (0–86400 sec): "))
category = input("Enter Merchant Category (e.g. Grocery, Travel, etc.): ")
age = int(input("Enter Customer Age: "))
gender = input("Enter Customer Gender (Male/Female/Other): ")
income = float(input("Enter Customer Income (INR): "))
location = input("Enter Transaction Location (City): ")
prev_frauds = int(input("Enter Previous Fraud Count: "))

#  Encode Categorical Features using previously fitted label encoders
category_encoded = label_encoders['Merchant Category'].transform([category])[0]
gender_encoded = label_encoders['Customer Gender'].transform([gender])[0]
location_encoded = label_encoders['Transaction Location'].transform([location])[0]

#  Create input DataFrame with same feature structure
user_input_knn = pd.DataFrame([[
    amount,
    time,
    category_encoded,
    age,
    gender_encoded,
    income,
    location_encoded,
    prev_frauds
]], columns=X.columns)

#  Scale user input (using same scaler from training phase)
user_input_knn_scaled = scaler.transform(user_input_knn)

#  Predict with KNN
knn_prediction = knn.predict(user_input_knn_scaled)[0]

# Output Result
print("\n🎯 Final KNN Model Prediction:")
print("🚨 Fraudulent Transaction ❌" if knn_prediction == 1 else "✅ Safe Transaction")
```
