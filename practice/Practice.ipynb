```python
##Linear regression
import numpy as np
import matplotlib.pyplot as plt

X=np.array([1,2,3,4,5,6])
Y=np.array([35,50,65,80,95,110])
print(X)
print(Y)

X_mean=np.mean(X)
Y_mean=np.mean(Y)
print(Y_mean)
print(X_mean)

#Calculations
#1. Slope(m)
numerator = np.sum((X-X_mean)*(Y-Y_mean))
denomenator= np.sum((X-X_mean)**2)

m=numerator/denomenator
print(m)

#Intercept(c)=Y-mX
c=Y_mean-m*X_mean
print(c)

Y_pred=m*X+c

plt.scatter(X, Y, color='red', label='Actual')
plt.plot(X, Y_pred, color='blue', label='Predicted Line')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Linear Regression: Years vs Salary")
plt.legend()
plt.grid()
plt.show()
```

```python
#checking MSE
mse=np.mean(Y-Y_mean)**2
print(f"Mean Squared Error: {mse}%")
```
